Importando das bibliotecas 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Carregando dos dados 

```python
### Carregando dados 
#--------------------------------------------------------------------------------------------------------------------------------------------------------
## Link da planilha do dataset
link ='https://github.com/carolaine67892/escola_filtros/blob/c7add213f7b3b7a1314e32b1b146c2c7b7f81bd0/aluno_filtro.csv?raw=True'
#--------------------------------------------------------------------------------------------------------------------------------------------------------
# Carregando o dataset
aluno = pd.read_csv(link)
#--------------------------------------------------------------------------------------------------------------------------------------------------------
# Mostrando o cabeçalho do dataset
aluno.head()
```

Análise e seleção das váriavéis 

```python
## Informações gerais das variáveis
display(aluno.info())
#--------------------------------------------------------------------------------------------------------------------------------
# Selecionando variáveis para investigação
df = aluno.loc[:, ['id','nome_escola_id', 'nu_ano_censo', 'id_aluno', 'id_matricula', 'nu_mes', 'nu_ano', 'nu_idade_referencia' , 'tp_sexo','tp_cor_raca' ]].dropna().reset_index(drop=True)
display(df.head())
#--------------------------------------------------------------------------------------------------------------------------------

```

Número de matriculados ao longo dos anos

```python
id_aluno = pd.DataFrame(df.groupby('nu_ano_censo')["id_aluno"].count())
id_aluno.reset_index(inplace = True)
id_aluno
```

##Filtro 
Para a análise ser mais detalhada focamos no ano de 2019.

```python

print(df['id_matricula'].value_counts())
#-------------------------------------------------------------------
# Aplicação do intervalo do filtro escolhido(a) pelo(a) discente
df_fil = df.loc[(df['nome_escola_id'] == 474) &    
                            (df['id_matricula'] > 210000000) &
                (df['nu_ano_censo'] == 2019), :].reset_index(drop=True)
#-------------------------------------------------------------------

display(df_fil)
```

```python
num_alunos = len(df_fil)
print(num_alunos)
```

Para análise gráfica foi utilizada a variável id_aluno ao invés de id_matrícula, isto porque ao realizar o gráfico, o mesmo apresentou diversos problemas, como o tipo de texto e decidiu-se utilizar id_aluno. 

```python
tp_sexo = pd.DataFrame(df_fil.groupby('tp_sexo')["id_aluno"].count())
tp_sexo.reset_index(inplace = True)
tp_sexo
```

```python
plt.pie(x = tp_sexo['id_aluno'], labels = tp_sexo['tp_sexo'], autopct = '%1.1f%%', startangle= 90)
plt.title("Pct(%) dos alunos por sexo na Colégio Estadual Evaristo da Veiga")
```
